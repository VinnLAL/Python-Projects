```python
#Question 1
import numpy as np
import scipy as sc
from matplotlib import pyplot as plt

def dydt(y,t):
    dydt = -2*y*t
    
    return dydt

y0 = 1
t = np.linspace(0,2,10)
y = sc.integrate.odeint(dydt, y0, t)

#Plotting graph
plt.figure()

plt.plot(t,y,color="red",label="y")
plt.title("Graph of y(t) versus (t)")
plt.xlabel("t")
plt.ylabel("y(t)")

plt.legend()
plt.show()
```

```python
#Question 2

import numpy as np
import scipy as sc

#In the form of AX = B
# [2 , 4 , 6][x] = [18]
# [5 , 3 , 2][y] = [13]
# [7 , 8 , 9][z] = [30]

# Defining coefficient matrix A and B using Numpy arrays
A = np.array([[2,4,6],[5,3,2],[7,8,9]])
B = np.array([[18],[13],[30]])


# Using the numpy.linalg.solve function to find the solution
X = np.linalg.solve(A,B)

#Display solution vector
print(X)

```

```python
# Question 3
from matplotlib import pyplot as plt
import requests

loop = True
while loop:
    day = input("Enter a valid day between 1-16: ")
    if (int(day) <= 16 and int(day) >= 1):
        loop = False
  


api_url = "https://api.open-meteo.com/v1/forecast?latitude=2.5&longitude=112.5&hourly=temperature_2m&timezone=Asia%2FSingapore&forecast_days=" + day
response = requests.get(api_url)

forecast_data = response.json()
    
# 24 hours of temperatures
temperature_list = forecast_data["hourly"]["temperature_2m"]
time = forecast_data["hourly"]["time"]

# Plotting graph
plt.figure()

plt.plot(time, temperature_list, color="red", label="Temperature")
plt.title(f"Temperature at Latitude: 2.375, Longitude: 112.5, Elevation: 272.0, Days: {day}")
plt.xlabel("Hours")
plt.ylabel("Temperature (Celsius)")

plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
